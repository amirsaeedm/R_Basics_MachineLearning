# Anova Test

```r
item1 <- c(22,42,44,52,45,37)
item2 <- c(52,33,8,47,43,32)
item3 <- c(16,24,19,18,34,39)

r1 <- c(item1, item2, item3)
r1
```

```r
tm1 <- rep(c('item1', 'item2', 'item3'), each=6)
tm1
```

```r
av1 = aov(r1 ~ tm1)  # (Values ~ Item_Mapping)
av1
```

```r
summary(av1)  #Fstats value = 2.541
```

- p-value of 0.11 > 0.05 significance level. Do not reject Null Hypothesis.
- Means that the mean sales volume of new menu items are equal.
