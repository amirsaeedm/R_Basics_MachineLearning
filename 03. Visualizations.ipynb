```r
p <- ggplot(airquality, aes (x = Month, y = Ozone)) +
      geom_boxplot (fill = "blue", colour = "black") +
      scale_y_continuous (name = "Mean ozone in nparts per billion",   ## Y Axis Title
      breaks = seq (0, 175,20)) +                                    ## Scale of Y Axis
      scale_x_discrete (name = "Monthwise") +                        ## X Axis Title
      ggtitle ("Boxplot of mean ozone by month")
p
```

# 7. Visualizations using ggplot2

```r
library(ggplot2)
```

```r
df <- data.frame(dose=c("D0", "D1", "D2"), len=c(4.2, 10, 29.5))
```

### 7.1 Basic BarPlot

```r
p <- ggplot(data=df, aes(x=dose, y=len)) +  
    geom_bar(stat="identity")
```

```r
class(p)
p
```

### 7.2 Horizontal BarPlot

```r
# Horizontal bar plot
#    ggplot(data=df, aes(x=dose, y=len)) +  
#    geom_bar(stat="identity")  +  
#    coord_flip()


p <- p + coord_flip()
p
```

### 7.3 Change width of BarPlot

```r
# Change the width of bars
ggplot(data=df, aes(x=dose, y=len)) +
  geom_bar(stat="identity", width= 0.5)
```

### 7.4 Change color of  BarPlot

```r
# Change colors
ggplot(data=df, aes(x=dose, y=len)) +
  geom_bar(stat="identity", color="blue", fill="lightgreen")
```

### 7.5 Change color witth minimal theme BarPlot

```r
# Minimal theme + blue fill color

ggplot(data=df, aes(x=dose, y=len)) +
  geom_bar(stat="identity", fill="steelblue") +
  theme_minimal()
```

### 7.6 Chose bars names to display BarPlot

```r
# Choose which items to display :

ggplot(data=df, aes(x=dose, y=len)) +
  geom_bar(stat="identity", fill="steelblue") +
  theme_minimal() + 
  scale_x_discrete(limits= c("D0", "D2"))
```

### 7.7 BarPlot Labels

```r
# Outside bars
ggplot(data=df, aes(x=dose, y=len)) +
  geom_bar(stat="identity", fill="steelblue") +
  geom_text(aes(label=len), vjust= -4, size=5) +
  theme_minimal()
```

```r
# Inside bars
ggplot(data=df, aes(x=dose, y=len)) +
  geom_bar(stat="identity", fill="steelblue") +
  geom_text(aes(label=len), vjust= 2, color="white", size=5) +
  theme_minimal()
```
